```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#x = np.linspace(1,2,20)
#y = np.linspace(2,3,20)
x = np.arange(1,2,0.05)
y = np.arange(2,3,0.05)

plt.figure(num=0,figsize=[15,5])
plt.subplot(131)
plt.plot(x,y,marker="x")
plt.title('Linear line')
#plt.show()


y = y + 1/2*np.random.rand(np.size(y))

#plt.figure(num=1,figsize=[5,5])
plt.subplot(132)
plt.plot(x,y,marker="x")
plt.title('Randomized')
plt.show()

w = 1/4*np.random.rand(1)
b = 1/4*np.random.rand(1)

#w = 2/np.sqrt(Ni+1)*(np.random.rand(Nh,Ni+1) - 0.5) # Xavier initialization
#b = 2/np.sqrt(Ni+1)*(np.random.rand(Nh,Ni+1) - 0.5) # Xavier initialization


alpha = 0.001

N = np.size(y)

input_ = x

for i in range(0,50):
    for k in range(0,N):
        output = np.dot(np.asarray([w,b]).T,np.asarray([input_[k],1]))
    
        dedw = input_[k]
    
        dLde = output - y[k]
    
        dedb = 1
    
        dLdw = dLde*dedw
    
        dLdb = dLde*dedb
    
        w = w - alpha*dLdw
    
        b = b - alpha*dLdb
    
#    plt.subplot(133)
    plt.plot(x,w*x+b,'-r', x,y,marker="x")
    plt.title('Regression by Gradient Descent method')
    plt.show()

print("weight: ", w)
print("bias: ", b)




return



```

## Linear Regression by Class

```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt

%matplotlib inline

class regression_linear:
    
    def __init__(self, inputnodes, outputnodes, learningrate):
        self.inodes = inputnodes
        self.onodes = outputnodes
        
        self.lr = learningrate
        
        self.w = np.random.rand(self.onodes,self.inodes)
        self.b = np.random.rand(self.onodes,self.inodes)
        
        pass
    
    def train(self, input_list, target_list):
        inputs = np.array(input_list, ndmin=2)
        targets = np.array(target_list, ndmin=2)
        
        output = np.dot(np.asarray([self.w[0],self.b[0]]).T,np.asarray([[inputs[0]],[1]]))
        
        dLde = output - targets[0] # vector
        
        dedw = inputs[0]
        
        dedb = 1
        
        dLdw = dLde*dedw # vector
        
        dLdb = dLde*dedb # vector
        
        self.w[0,0] += - self.lr*dLdw[0]
        
        self.b[0,0] += - self.lr*dLdb[0]
        
        pass
    
    def query(self):
        print('Hello !')
        print(self.w)
        print(self.b)
        
        return self.w[0,0], self.b[0,0]
        pass
        
        
```

```python
input_nodes = 1
output_nodes = 1

learning_rate = 0.001

R = regression_linear(input_nodes, output_nodes, learning_rate)
#R = regression_linear(1, 1, 0.001)
R.query()
#R.query

#====================================================
x = np.arange(1,2,0.05)
y = np.arange(2,3,0.05)

plt.figure(num=0,figsize=[15,5])
plt.subplot(131)
plt.plot(x,y,marker="x")
plt.title('Linear line')
#plt.show()


y = y + 1/2*np.random.rand(np.size(y))

#plt.figure(num=1,figsize=[5,5])
plt.subplot(132)
plt.plot(x,y,marker="x")
plt.title('Randomized')
plt.show()
#====================================================

N = np.size(y)

input_ = x

for i in range(0,50):
    for k in range(0,N):
        R.train(input_[k], y[k])
        
    plt.plot(x,R.query()[0]*x + R.query()[1], '-r', x,y,marker="x")   
    plt.title('Regression by GDM')
    plt.show()


```
